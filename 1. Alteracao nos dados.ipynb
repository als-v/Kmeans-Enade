# Classificador K-means com os dados do ENADE

## Alterar valores para se utilizar o K-means

### 1. Inicialmente abrir o arquivo 'data/enade_merge_08-11-14-17.csv'

```python
import pandas as pd
```

```python
data = pd.read_csv('data/enade_merge_08-11-14-17.csv', sep=',')

data
```

### 2. Informações sobre os dados

```python
data.info()
```

### 3. Remoção das colunas que não serão utilizadas

```python
dropColumns = ['curso', 'codigo_IES', 'organizacao_academica', 'ead_presencial', 'formatura_ensino_medio', 'inicio_graduacao', 'range_idade', 'renda_familiar', 'cotas', 'bolsa_academica', 'nota_formacao_geral', 'nota_obj_formacao_geral', 'nota_dis_formacao_geral', 'nota_componente_especifico', 'nota_obj_componente_especifico', 'nota_dis_componente_especifico']
data.drop(dropColumns, axis=1, inplace=True)

data.info()
```

### 4. Alteração dos atributos numéricos

```python
data.loc[data['sexo'] == 'M', 'sexo'] = 1
data.loc[data['sexo'] == 'F', 'sexo'] = 0

data['sexo'].value_counts()
```

```python
data.loc[data['recebeu_cota'] == 'Sim', 'recebeu_cota'] = 1
data.loc[data['recebeu_cota'] == 'Não', 'recebeu_cota'] = 0

data['recebeu_cota'].value_counts()
```

```python
data.loc[data['recebeu_bolsa'] == 'Sim', 'recebeu_bolsa'] = 1
data.loc[data['recebeu_bolsa'] == 'Não', 'recebeu_bolsa'] = 0

data['recebeu_bolsa'].value_counts()
```

```python
# os valores: 'Não' e 'Não-Informado' serão o mesmo
data.loc[data['trabalha'] == 'Sim', 'trabalha'] = 1
data.loc[
    (data['trabalha'] == 'Não') | 
    (data['trabalha'] == 'Não-Informado'),
    'trabalha'
] = 0

data['trabalha'].value_counts()
```

```python
data.loc[data['livros_ano'] == '0', 'livros_ano'] = 0
data.loc[data['livros_ano'] == '1/2', 'livros_ano'] = 1
data.loc[data['livros_ano'] == '3-5', 'livros_ano'] = 2
data.loc[data['livros_ano'] == '6-8', 'livros_ano'] = 3
data.loc[data['livros_ano'] == '8+', 'livros_ano'] = 4

data['livros_ano'].value_counts()
```

```python
data.loc[data['horas_estudo'] == '0', 'horas_estudo'] = 0
data.loc[data['horas_estudo'] == '<3', 'horas_estudo'] = 1
data.loc[data['horas_estudo'] == '3-8', 'horas_estudo'] = 2
data.loc[data['horas_estudo'] == '8+', 'horas_estudo'] = 3

data['horas_estudo'].value_counts()
```

```python
data.loc[
    (data['pessoas_moradia'] == 'Não-Informado') | 
    (data['pessoas_moradia'] == 'Outro'),
    'pessoas_moradia'
] = 0

data.loc[data['pessoas_moradia'] == 'Nenhuma', 'pessoas_moradia'] = 1
data.loc[data['pessoas_moradia'] == 'Uma', 'pessoas_moradia'] = 2
data.loc[data['pessoas_moradia'] == 'Duas', 'pessoas_moradia'] = 3
data.loc[data['pessoas_moradia'] == 'Três', 'pessoas_moradia'] = 4
data.loc[data['pessoas_moradia'] == 'Quatro', 'pessoas_moradia'] = 5
data.loc[data['pessoas_moradia'] == 'Cinco', 'pessoas_moradia'] = 6
data.loc[data['pessoas_moradia'] == 'Seis', 'pessoas_moradia'] = 7
data.loc[data['pessoas_moradia'] == 'Mais de seis', 'pessoas_moradia'] = 8

data['pessoas_moradia'].value_counts()
```

```python
data.loc[
    (data['condicoes_salas_estudantes'] == 'Não-Informado') | 
    (data['condicoes_salas_estudantes'] == 'Outro') |
    (data['condicoes_salas_estudantes'] == 'Não se aplica'),
    'condicoes_salas_estudantes'
] = 0

data.loc[data['condicoes_salas_estudantes'] == 'Todas', 'condicoes_salas_estudantes'] = 4
data.loc[data['condicoes_salas_estudantes'] == 'Maior parte', 'condicoes_salas_estudantes'] = 3
data.loc[data['condicoes_salas_estudantes'] == 'Algumas', 'condicoes_salas_estudantes'] = 2
data.loc[data['condicoes_salas_estudantes'] == 'Nenhuma', 'condicoes_salas_estudantes'] = 1

data['condicoes_salas_estudantes'].value_counts()
```

```python
data.loc[
    (data['condicoes_praticas_materiais'] == 'Não-Informado') | 
    (data['condicoes_praticas_materiais'] == 'Outro') |
    (data['condicoes_praticas_materiais'] == 'Não se aplica'),
    'condicoes_praticas_materiais'
] = 0

data.loc[data['condicoes_praticas_materiais'] == 'Todas', 'condicoes_praticas_materiais'] = 4
data.loc[data['condicoes_praticas_materiais'] == 'Maior parte', 'condicoes_praticas_materiais'] = 3
data.loc[data['condicoes_praticas_materiais'] == 'Algumas', 'condicoes_praticas_materiais'] = 2
data.loc[data['condicoes_praticas_materiais'] == 'Nenhuma', 'condicoes_praticas_materiais'] = 1

data['condicoes_praticas_materiais'].value_counts()
```

```python
data.loc[
    (data['plano_ensino'] == 'Não-Informado') | 
    (data['plano_ensino'] == 'Outro') |
    (data['plano_ensino'] == 'Não se aplica'),
    'plano_ensino'
] = 0

data.loc[data['plano_ensino'] == 'Todos aspectos', 'plano_ensino'] = 4
data.loc[data['plano_ensino'] == 'Maior parte', 'plano_ensino'] = 3
data.loc[data['plano_ensino'] == 'Somente alguns', 'plano_ensino'] = 2
data.loc[data['plano_ensino'] == 'Nenhum', 'plano_ensino'] = 1

data['plano_ensino'].value_counts()
```

```python
data.loc[
    (data['avaliacao_curso'] == 'Não-Informado') | 
    (data['avaliacao_curso'] == 'Outro') |
    (data['avaliacao_curso'] == 'Não se aplica'),
    'avaliacao_curso'
] = 0

data.loc[data['avaliacao_curso'] == 'Muito boa', 'avaliacao_curso'] = 5
data.loc[data['avaliacao_curso'] == 'Boa', 'avaliacao_curso'] = 4
data.loc[data['avaliacao_curso'] == 'Regular', 'avaliacao_curso'] = 3
data.loc[data['avaliacao_curso'] == 'Fraca', 'avaliacao_curso'] = 2
data.loc[data['avaliacao_curso'] == 'Muito fraca', 'avaliacao_curso'] = 1

data['avaliacao_curso'].value_counts()
```

```python
anos = [2008, 2011, 2014, 2017]

for ano in anos:
    dados = data[data['ano'] == ano]
    dados.to_csv(f'./data/not_norm/dados{ano}_nn.csv', index=False)
```
